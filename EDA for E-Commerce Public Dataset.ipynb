### Cleaning Data

#### 1. Pembersihan Data pada `customers_dataset.csv`

##### a. Mengubah huruf pertama dari setiap kata di kolom `customer_city` pada tabel `df_customers` menjadi huruf kapital.

```python
# Fungsi untuk mengubah huruf awal setiap kata
def capitalize_city(city):
    return " ".join([word.capitalize() for word in city.split()])

# Menerapkan fungsi pada kolom "customer_city" pada tabel df_customers
df_customers["customer_city"] = df_customers["customer_city"].apply(capitalize_city)

# Tampilkan hasil
df_customers.head(2)
```

##### b. Mengganti singkatan dalam kolom `customer_state` pada tabel `df_customers` dengan kepanjangan nama yang bersumber dari situs https://www.drupal.org/node/2559551.

```python
# Kamus pemetaan singkatan ke kepanjangan nama dari situs https://www.drupal.org/node/2559551
state_mapping = {
    'AC': 'Acre',
    'AL': 'Alagoas',
    'AP': 'Amap',
    'AM': 'Amazonas',
    'BA': 'Bahia',
    'CE': 'Cear',
    'DF': 'Distrito Federal',
    'ES': 'Esprito Santo',
    'GO': 'Gois',
    'MA': 'Maranho',
    'MT': 'Mato Grosso',
    'MS': 'Mato Grosso do Sul',
    'MG': 'Minas Gerais',
    'PA': 'Par',
    'PB': 'Paraba',
    'PR': 'Paran',
    'PE': 'Pernambuco',
    'PI': 'Piau',
    'RJ': 'Rio de Janeiro',
    'RN': 'Rio Grande do Norte',
    'RS': 'Rio Grande do Sul',
    'RO': 'Rondnia',
    'RR': 'Roraima',
    'SC': 'Santa Catarina',
    'SP': 'So Paulo',
    'SE': 'Sergipe',
    'TO': 'Tocantins'
}

# Menggantikan singkatan dengan kepanjangan nama pada kolom "customer_state"
df_customers['customer_state'] = df_customers['customer_state'].map(state_mapping)

# Tampilkan hasil
df_customers.head(2)
```

#### 2. Pembersihan Data pada `order_reviews_dataset.csv`

##### a. Menghapus kolom `review_comment_title` dan `review_comment_message` karena tidak digunakan dalam proses EDA dan Analisis

```python
# Menghapus kolom 'review_comment_title' dan 'review_comment_message'
df_order_reviews = df_order_reviews.drop(['review_comment_title', 'review_comment_message'], axis=1)
```

##### b. Mengganti kolom `review_creation_date` dan `review_answer_timestamp` menjadi tipe data datetime

```python
# Mengganti kolom 'review_creation_date' dan 'review_answer_timestamp' menjadi tipe data datetime
df_order_reviews['review_creation_date'] = pd.to_datetime(df_order_reviews['review_creation_date'])
df_order_reviews['review_answer_timestamp'] = pd.to_datetime(df_order_reviews['review_answer_timestamp'])
```

```python
df_order_reviews
```

#### 3. Pembersihan Data pada `orders_dataset.csv`

##### a. Merubah label kolom tanggal yang akan digunakan nanti agar lebih singkat dan sederhana.

```python
# Mengganti nama kolom
df_orders = df_orders.rename(columns={
    "order_purchase_timestamp": "order_date",
    "order_approved_at": "approved_date",
    "order_delivered_carrier_date": "shipped_date",
    "order_delivered_customer_date": "delivery_date",
    "order_estimated_delivery_date": "estimated_date"
})

# Tampilkan DataFrame setelah perubahan nama kolom
df_orders.head(2)
```

##### b. Mengubah tipe data beberapa kolom menjadi datetime

```python
# Kolom tanggal yang akan diubah menjadi tipe data datetime
date_columns = ['order_date', 'approved_date', 'shipped_date', 'delivery_date', 'estimated_date']

# Mengubah kolom-kolom tersebut menjadi tipe data datetime
df_orders[date_columns] = df_orders[date_columns].apply(pd.to_datetime)

# Tampilkan DataFrame setelah perubahan
df_orders.head(2)
```

```python
# Pastikan kolom 'order_date' sudah dalam format datetime
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])

# Insert kolom 'hour_order' di index ke-3
df_orders.insert(3, 'hour_order', df_orders['order_date'].dt.hour)

# Membagi kolom 'hour_order' ke dalam beberapa bagian
hours_bins = [-0.1, 6, 12, 18, 23]
hours_labels = ['Early Morning', 'Morning', 'Afternoon', 'Night']

df_orders.insert(4, 'daytime_order', pd.cut(df_orders['hour_order'], bins=hours_bins, labels=hours_labels))
df_orders.insert(5, 'day_order', df_orders['order_date'].dt.strftime('%A'))
df_orders.insert(6, 'date_order', df_orders['order_date'].dt.strftime('%d'))
df_orders.insert(7, 'month_order', df_orders['order_date'].dt.strftime('%m-%Y'))
df_orders.head(5)
```

```python
df_orders.info()
```

#### 6. Menggabungkan Tabel `product_category_name_translation.csv` dan `products.csv` menjadi **`merged_product_df`**

### Analisis RFM

RFM Analysis (Recency, Frequency, Monetary) adalah metode analisis pelanggan yang umumnya digunakan untuk mengidentifikasi dan memahami perilaku pembelian pelanggan. Mari kita aplikasikan konsep RFM pada data dan informasi yang telah kita diskusikan selama dua hari ini.

1. **Recency (Ketepatan Waktu):**
   - Dalam konteks pembelian, kita dapat mengidentifikasi pelanggan yang melakukan transaksi terakhirnya. Sebagai contoh, kita bisa melihat data terkini untuk menentukan pelanggan yang baru-baru ini melakukan pembelian. Data ini dapat digunakan untuk menilai seberapa "baru" atau "terkini" suatu pembelian.

2. **Frequency (Frekuensi):**
   - Dengan melibatkan informasi tentang jumlah pembelian atau transaksi yang dilakukan oleh pelanggan, kita dapat mengidentifikasi pelanggan yang sering berbelanja. Informasi ini dapat diambil dari data jumlah pembelian atau frekuensi pembelian selama periode waktu tertentu.

   Sudah dijelaskan pada bagian:
   - Melihat Id pelanggan yang paling sering dan jarang berbelanja berdasarkan `customer_city` dan `customer_state`.

3. **Monetary (Moneter):**
   - Dengan mengintegrasikan data terkait nilai pembelian atau pendapatan yang dihasilkan dari setiap transaksi, kita dapat mengidentifikasi pelanggan yang memberikan kontribusi signifikan terhadap pendapatan perusahaan. Ini melibatkan perhitungan nilai moneter total yang dihabiskan oleh pelanggan.
   
   Sudah dijelaskan pada bagian:
     - Melihat Id pelanggan dengan pendapatan tertinggi dan terendah berdasarkan `customer_city` dan `customer_state`.

Penerapan RFM Analysis pada data pembelian dan penjualan dapat membantu perusahaan dalam:
- Mengetahui pelanggan yang sering berbelanja dan memberikan kontribusi besar terhadap pendapatan.
- Mengklasifikasikan pelanggan ke dalam segmen atau kelompok berdasarkan perilaku pembelian mereka.
- Merancang strategi pemasaran yang lebih terfokus dan disesuaikan dengan kebutuhan setiap kelompok pelanggan.

Dalam konteks data dan informasi yang telah dibahas, penerapan RFM Analysis memungkinkan perusahaan untuk lebih memahami dan merespons secara efektif terhadap pola pembelian pelanggan. Dengan informasi ini, strategi pemasaran dan layanan pelanggan dapat dioptimalkan untuk meningkatkan retensi pelanggan dan hasil keseluruhan bisnis.
